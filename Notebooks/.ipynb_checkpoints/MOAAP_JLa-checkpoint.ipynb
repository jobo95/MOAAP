```python
import matplotlib.pyplot as plt 
import cartopy
import cartopy.crs as ccrs
import glob

%run ./Tracking_Functions.py

### Load input data
ds=xr.open_mfdataset('/work/aa0049/a271109/data_misc/ERA5/data_processed/MSL_ERA5_2020-04_hourly_remapped.nc')
ds=ds.sel(lat=slice(50,None))

Time=np.array(pd.DataFrame(ds.time.values,columns=['Datetime'])['Datetime'].values.astype('datetime64[s]').tolist())
Lon, Lat= xr.broadcast(ds.lon, ds.lat)[0].values.T, xr.broadcast(ds.lon, ds.lat)[1].values.T[:,:]
Mask=1*(Lat>50) 

### run MOAAP
moaap(Lon = Lon,                           # 2D longitude grid centers
      Lat = Lat,                           # 2D latitude grid spacing
      Time = Time,                         # datetime vector of data
      dT = 1,                              # integer - temporal frequency of data [hour]
      Mask = Mask,                         # mask with dimensions [lat,lon] defining analysis region
      
      slp = ds.MSL,                        # sea level pressure [Pa]

      DataName = 'ERA5_cyclones',
      OutputFolder='./output/'
     )  

### Plot output example
ds_out=xr.open_dataset('./output/202004_ERA5_cyclones_ObjectMasks_ERA5_cyclones_dt-1h_MOAAP-masks.nc')
#ds_out

def plot_AR(timsel=None):
    fig,ax=plt.subplots(1,1,subplot_kw={'projection': ccrs.Orthographic(0,90)}, figsize=(6,6),facecolor='w')  
    
    ds_out.SLP.sel(time=timsel).plot(x='lon',y='lat',ax=ax,transform=ccrs.PlateCarree())

    (ds_out.CY_Objects.sel(time=timsel)>0).plot.contour(x='lon',y='lat',levels=1,ax=ax,transform=ccrs.PlateCarree(),colors='k',linewidths=1) 
    (ds_out.ACY_Objects.sel(time=timsel)>0).plot.contour(x='lon',y='lat',levels=1,ax=ax,transform=ccrs.PlateCarree(), colors='r',linewidths=1) 
    
    for aa in [ax]:
        aa.set_extent([-180,180, 50, 90], crs=ccrs.PlateCarree())
        aa.add_feature(cartopy.feature.OCEAN, color='white', zorder=0)
        aa.add_feature(cartopy.feature.LAND, color='lightgray',zorder=0,linewidth=0.5, edgecolor='black')
        aa.gridlines(linewidth=0.5, color='gray',xlocs=range(-180,180,45), ylocs=range(-90,90,10)) #draw_labels=True, 
        aa.coastlines(resolution='50m', linewidth=0.3, color='black')
    
plot_AR(timsel='2020-04-17T12:00:00.000000000')
```
