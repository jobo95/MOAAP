```python
%load_ext autoreload
%autoreload 2

from sklearn import decomposition, cluster
import xarray as xr
import os
from src.clustering_utils import *
```

```python
year_start = 1979
year_end = 2018
slp_nc = "MSL"
lon_nc = 'lon'  
lat_nc = 'lat' 
time_nc = 'time' 

path_in = '/work/aa0238/a271093/data/ERA5/1979-2023/original_data/slp/daily/'
name_file_slp = 'ERA5_slp.4dx4dy.N_all_del29feb_-90_90_20_88_aac.nc'  # Area selected,  29feb deleted, all months !!!!!!

DO_reduce_dim = 1  #reduce dimensionality before applying clustering
DO_cosine_cor = 1     #DO cosine correction to data before applying clustering 
                      #and transorm back afterwards


if DO_reduce_dim:
    n_components = 20  #dimensionality of reduced phase space

##################
#kmeans parameter#
##################
num_cluster = 5
num_iter = 50     # initiate algorithm n_times randomly and pick best result
random_state = 42   #reandom seed for reproducibility
```

```python

# get slp_data
slp_data,lat,lon= get_2d_field(path_in,
                               name_file_slp,
                               slp_nc,
                               [12,1,2],
                               year_start,
                               year_end,
                               lon_nc,
                               lat_nc,
                               time_nc)

time_frame = slp_data.index
slp_data = slp_data.values
```

```python

if DO_cosine_cor:
    slp_data =  cosine_cor(slp_data,lat,lon)
    
if DO_reduce_dim:
    pca = decomposition.PCA(n_components=n_components)
    pca.fit(slp_data)
    slp_data = pca.transform(slp_data)
```

```python
#####################
#######          ####
###### KMeans    ####
#####################


means = cluster.KMeans(n_clusters=num_cluster,
                n_init=num_iter,
                max_iter=900 ,
                random_state = random_state,
                algorithm='lloyd').fit(slp_data)
cluster_centers = means.cluster_centers_
plot_len = num_cluster
BMU = means.labels_
    
```

```python
if DO_reduce_dim:
    cluster_centers = pca.inverse_transform(cluster_centers)
```

```python
if DO_cosine_cor:
    cluster_centers = inv_cosine_cor(cluster_centers,lat,lon)
    
```

```python


    
plot_dat = cluster_centers.reshape(-1,lat.size,lon.size)
cluster_names = np.arange(5)

fig_cluster = plot_cluster(plot_dat,lat,lon,
                            plot_size=(35, 15),
                            plot_shape=[1,5],
                            cbar_size=1,
                            cbar_ticks=[-15, -10, -8, -3, -2,-1,1, 2, 3, 8, 10, 15],
                            font_size=17,
                            unit='hPa',
                            color_lev = [-15, -10, -8, -3, -2,-1,1, 2, 3, 8, 10, 15],
                            titles =np.arange(5))
```
