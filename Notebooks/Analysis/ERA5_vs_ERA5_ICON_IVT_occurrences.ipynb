# Plot number of days per season where a grid cell is part of an IVT object

```python
fig = plt.figure(figsize=(13,10))
subplts = (3,4)
levels=np.arange(1,30,1)
levels=[1,2,3,5,7,9,12,15,20,25,30]
for i, season_ in enumerate(Season):
    print (season_)
        
    IVTobjs_sel_ERA5 = IVT_objs_ERA5.sel_season(season_)

    lon,lat,z_ERA5 = count_objs_grid_points(IVTobjs_sel_ERA5, normalization_factor=24.)
    
    # normalize grid point counts to counts per season
    z_ERA5= (z_ERA5/num_years)*1.001
    
    plot_unstructured_rotated_grid(lon,
                                  lat, 
                                  z_ERA5, 
                                  fig=fig,
                                  index = i,
                                  subplts=subplts,
 
                                  levels = levels,
                                  #title=f"ERA5 ({first_year}-{last_year}): {season_.name}",
                                  cbar_label = "IVT object days per season",
                                   cbar=True
                                  )
    

    IVTobjs_sel_ICON_ERA5 = IVT_objs_ICON_ERA5.sel_season(season_)

    lon,lat,z_ICON_ERA5 = count_objs_grid_points(IVTobjs_sel_ICON_ERA5, normalization_factor=24.)
    
    # normalize grid point counts to counts per season
    z_ICON_ERA5= (z_ICON_ERA5/num_years)*1.001


    plot_unstructured_rotated_grid(lon,
                                  lat, 
                                  z_ICON_ERA5, 
                                  fig=fig,
                                  index = i+4,
                                  subplts=subplts,
 
                                  levels = levels,
                                  title=f"ERA5 ({first_year}-{last_year}): {season_.name}",
                                  cbar_label = "IVT object days per season",
                                   cbar=False
                                  )
    
    plot_unstructured_rotated_grid(lon,
                                  lat, 
                                  z_ICON_ERA5-z_ERA5, 
                                  fig=fig,
                                  index = i+8,
                                  subplts=subplts,
                                   cmap="seismic",
 
                                  levels = np.arange(-1.8,1.9,0.3),
                                  #title=f"ERA5 minus ICON-ERA5:({first_year}-{last_year}): {season_.name}",
                                  cbar_label = "IVT object day difference per season",
                                   cbar=True
                                  )

#    plot_unstructured_rotated_grid(lon,
#                                  lat, 
#                                    100*(z_ICON_ERA5-(z_ERA5+0.1))/(z_ERA5+0.1),
#                                   fig=fig,
#                                  index = i+8,
#                                  subplts=subplts,
#                                   cmap="seismic",
# 
#                                  levels = np.arange(-35,36,5),
#                                  #title=f"ERA5 minus ICON-ERA5:({first_year}-{last_year}): {season_.name}",
#                                  cbar_label = "Relative increase in IVT object days [%]",
#                                   cbar=True
#                                  )
plt.tight_layout()
```
